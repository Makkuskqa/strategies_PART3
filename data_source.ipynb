```python
import requests
import pandas as pd
from datetime import datetime, timedelta
```

```python
end_date = datetime.today()
start_date = end_date - timedelta(days=1815)
symbol = 'NDX.INDX'
api_key = '657045d7ecbf83.00486901'
```

```python
request_url = f'https://eodhd.com/api/eod/{symbol}?from={start_date}&to={end_date}&api_token={api_key}&fmt=json'
data = requests.get(request_url)
```

```python
data.json()
```

```python
df = pd.DataFrame(data.json())
```

```python
df.head()
```

```python
df = df.drop(columns=['adjusted_close'])
df.to_csv('NDX5years.csv', index=False)
```
